# Sleep Duration Patterns
- Sleep Duration Distribution:

 The histogram suggests that sleep duration is distributed across a range, with a peak between 4 to 5 hours and another noticeable increase around 9 to 11 hours.

 This could indicate two subgroups: one group sleeping fewer hours (possibly due to lifestyle or work habits) and another group sleeping longer hours (likely more aligned with recommended sleep patterns).
- Density Curve (KDE)

The KDE curve highlights the underlying probability distribution of the sleep durations. It shows a smooth rise around 4–5 hours and 9–11 hours, with a dip around 7 hours. This suggests that sleep durations of 7 hours are less common among the population.
- Observations:

The sleep duration data may have two modes (bimodal distribution), indicating distinct behavior or characteristics within the dataset.

Frequency-wise, the most common durations are around 4–5 hours and 9–11 hours, with fewer individuals reporting a duration near 6–8 hours.

```python
sns.set(style="whitegrid")

plt.figure(figsize=(8, 6))
sns.histplot(df['Sleep Duration (hours)'], kde=True, palette="tab10")
plt.title('Distribution of Sleep Duration', fontsize=16, weight='bold', color='darkblue')
plt.xlabel('Sleep Duration (hours)', fontsize=12, weight='bold')
plt.ylabel('Frequency', fontsize=12, weight='bold')
plt.show()
```

# How does sleep quality relate to the amount of sleep

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x = "Sleep Duration (hours)" ,
                y = 'Quality of Sleep (scale: 1-10)', data=df )
plt.title ("Relation between sleep durationg and sleep quality")
plt.xlabel('Sleep Duration (hours)')
plt.ylabel('Quality of Sleep (scale: 1-10)')
plt.show()
```

# Is there any significant relationship between stress level and sleep

```python
plt.figure(figsize=(8,6))
sns.boxplot(x = "Sleep Disorder" , y = "Stress Level (scale: 1-10)",
            data = df,
            palette="Set2" )
plt.title('Stress Levels Across Different Sleep Disorders')
plt.xlabel('Sleep Disorder')
plt.ylabel('Stress Level')
plt.show()
```

```python

plt.figure(figsize=(8,6))
sns.kdeplot(data=df, x="Stress Level (scale: 1-10)", hue="Sleep Disorder", fill=True, common_norm=False, palette="Set2")

plt.title('KDE of Stress Levels Across Different Sleep Disorders')
plt.xlabel('Stress Level (scale: 1-10)')
plt.ylabel('Density')
plt.show()
```

#  What are the common patterns of sleep disorders in different age groups?

```python
age_bins = pd.cut(df['Age'], bins=[0, 18, 30, 40, 50, 60, 100], labels=['0-18', '19-30', '31-40', '41-50', '51-60', '60+'])
df['Age Group'] = age_bins
plt.figure(figsize=(8, 6))
sns.countplot(x='Age Group', hue='Sleep Disorder', data=df,palette="pastel")
plt.title('Prevalence of Sleep Disorders Across Age Groups')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.show()
```
